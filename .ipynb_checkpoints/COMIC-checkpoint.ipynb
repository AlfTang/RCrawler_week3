##  User Agent

```r
res = httr::GET("http://www.google.com")
```

```r
IRdisplay::display_html(content(res, type = "text", encoding = "big5"))
```

```r
res2 = GET("http://www.yungching.com.tw/")
```

```r
content(res2, type = "text", encoding = "utf8")
```

```r
res2["headers"]
```

```r
add_headers("User-Agent" = "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/43.0.2357.134 Safari/537.36")
```

```r
res3 = GET("http://www.yungching.com.tw/", add_headers("User-Agent" = "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/43.0.2357.134 Safari/537.36"))
```

```r
display_html(content(res3, type = "text", encoding = "utf8"))
```

```r
res3["headers"]
```

```r
res3$request$headers
```

# 實價登入- cookie 驗證碼

```r
library(httr)
library(jpeg)
library(CSS)
library(XML)
```

```r
res7 = GET("http://lvr.land.moi.gov.tw/N11/homePage.action")
cookies(res7)
```

```r

res8 = GET("http://lvr.land.moi.gov.tw/N11/ImageNumberN13?", add_c)
captcha <- content(res8)
plot(0:1, 0:1, type = "n")
rasterImage(captcha, 0, 0, 1, 1)
```

```r
res10 <- httr::POST('http://lvr.land.moi.gov.tw/N11/login.action', set_cookies(.cookies = cookies(res8)))
```

```r
display_html(content(res10, 'text'))
```
