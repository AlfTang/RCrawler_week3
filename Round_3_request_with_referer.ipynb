### 萬年起手式 - httr:: GET()

```r
library(httr)
cookie <- set_cookies(
    ' PHPSESSID'='15244387_dc9353b9f969b83f657db3a5f4eda801'
)
response.before <- httr::GET("http://i3.pixiv.net/img23/img/scaji/2362526.jpg", cookie, timeout(30))
```

## 來欣賞一下美圖吧 >////<

```r
library(IRdisplay)
display_jpeg(content(response.before, 'raw'))
```

## 讓我們再次檢查一下 header: 是你！ Referer ! 
* 再次請出 httr:: add_headers()
<a href='http://www.pixiv.net/member_illust.php?mode=medium&illust_id=2362526'><img src="fig/referer.png" height=430></a>

```r
header <- add_headers(
    "Referer"="http://www.pixiv.net/member_illust.php?mode=medium&illust_id=2362526"
)
response.after <- httr::GET("http://i3.pixiv.net/img23/img/scaji/2362526.jpg", cookie, header)
pic <- content(response.after, 'raw')
display_jpeg(pic)
```
